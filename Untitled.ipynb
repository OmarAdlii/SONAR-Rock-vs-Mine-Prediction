```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import f1_score ,jaccard_score ,accuracy_score ,log_loss ,confusion_matrix
```

```python
dir=os.path.join('data.csv')
data=pd.read_csv(dir,header=None)
data.shape
```

```python
data.groupby(60).mean()
```

```python
X=data.drop(columns=60,axis=1)
y=data[60].copy()
y.replace(['R', 'M'], [0,1], inplace=True)
```

```python
X
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=42,stratify=y)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
LR_model=LogisticRegression().fit(x_train,y_train)
KNN_model=KNeighborsClassifier(n_neighbors=5).fit(x_train,y_train)
DTree_model=DecisionTreeClassifier(criterion='entropy',max_depth=10).fit(x_train,y_train)
SVM_model=SVC(kernel='linear').fit(x_train,y_train)
```

```python
LR_predictions=LR_model.predict(x_test)
KNN_predictions=KNN_model.predict(x_test)
Dtree_predictions=DTree_model.predict(x_test)
SVM_predictions=SVM_model.predict(x_test)
```

```python
methods = []
results=[]
```

```python
LR_Accuracy_Score = accuracy_score(y_test,LR_predictions)
LR_JaccardIndex = jaccard_score(y_test,LR_predictions)
LR_F1_Score = f1_score(y_test,LR_predictions)
LR_Log_Loss = log_loss(y_test, LR_predictions)
methods.append('logistic regression')
results.append([LR_Accuracy_Score,LR_JaccardIndex,LR_F1_Score,LR_Log_Loss])
print('LR_Accuracy_Score = %.2f'%LR_Accuracy_Score)
print('LR_JaccardIndex = %.2f'%LR_JaccardIndex)
print('LR_F1_Score = %.2f'%LR_F1_Score)
print(f'LR_Log_Loss = %.2f'%LR_Log_Loss)
```

```python
KNN_Accuracy_Score = accuracy_score(y_test,KNN_predictions)
KNN_JaccardIndex = jaccard_score(y_test,KNN_predictions)
KNN_F1_Score = f1_score(y_test,KNN_predictions)
KNN_Log_Loss = log_loss(y_test, KNN_predictions)
methods.append('K-nearest neighbors')
results.append([KNN_Accuracy_Score,KNN_JaccardIndex,KNN_F1_Score,KNN_Log_Loss])
print('KNN_Accuracy_Score = %.2f'%KNN_Accuracy_Score)
print('KNN_JaccardIndex = %.2f'%KNN_JaccardIndex)
print('KNN_F1_Score = %.2f'%KNN_F1_Score)
print(f'KNN_Log_Loss = %.2f'%KNN_Log_Loss)
```

```python
DT_Accuracy_Score = accuracy_score(y_test,Dtree_predictions)
DT_JaccardIndex = jaccard_score(y_test,Dtree_predictions)
DT_F1_Score = f1_score(y_test,Dtree_predictions)
DT_Log_Loss = log_loss(y_test, Dtree_predictions)
methods.append('decision tree classifier')
results.append([DT_Accuracy_Score,DT_JaccardIndex,DT_F1_Score,DT_Log_Loss])
print('DT_Accuracy_Score = %.2f'%DT_Accuracy_Score)
print('DT_JaccardIndex = %.2f'%DT_JaccardIndex)
print('DT_F1_Score = %.2f'%DT_F1_Score)
print(f'DT_Log_Loss = %.2f'%DT_Log_Loss)
```

```python
SVM_Accuracy_Score = accuracy_score(y_test,SVM_predictions)
SVM_JaccardIndex = jaccard_score(y_test,SVM_predictions)
SVM_F1_Score = f1_score(y_test,SVM_predictions)
SVM_Log_Loss = log_loss(y_test, SVM_predictions)
methods.append('support vector machine')
results.append([SVM_Accuracy_Score,SVM_JaccardIndex,SVM_F1_Score,SVM_Log_Loss])
print('SVM_Accuracy_Score = %.2f'%SVM_Accuracy_Score)
print('SVM_JaccardIndex = %.2f'%SVM_JaccardIndex)
print('SVM_F1_Score = %.2f'%SVM_F1_Score)
print(f'SVM_Log_Loss = %.2f'%SVM_Log_Loss)
```

```python
def plot_confusion_matrix(y,y_predict,model):
   
    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax);
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix ---> ('+model+')'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

```python
plot_confusion_matrix(y_test,LR_predictions,'Linear Logistic model')
```

```python
plot_confusion_matrix(y_test,KNN_predictions,'KNN model')
```

```python
plot_confusion_matrix(y_test,SVM_predictions,'SVM model')
```

```python
plot_confusion_matrix(y_test,Dtree_predictions,'Decision Tree model')
```

```python
fig = plt.figure(figsize = (10, 6))
plt.bar(methods,[row[0] for row in results])
plt.xlabel("Methods")
plt.ylabel("Accuracy")
plt.title("Comparison of methods")
plt.show()
```
